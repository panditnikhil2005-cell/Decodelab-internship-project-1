4. Delete tasks
5. Exit the program

## Algorithm / Working
1. Start the program.
2. Create an empty list to store tasks.
3. Show a menu with options:
   - Add Task
   - View Tasks
   - Mark Task as Completed
   - Delete Task
   - Exit
4. If the user selects **Add Task**, take task name as input and store it in the list with status **Pending**.
5. If the user selects **View Tasks**, display all tasks with their status.
6. If the user selects **Mark Task as Completed**, ask for task number and update its status to **Completed**.
7. If the user selects **Delete Task**, ask for task number and remove it from the list.
8. If the user selects **Exit**, stop the program.


````python

# =========================
# TO-DO LIST APPLICATION
# =========================

tasks = []

while True:
    print("\n========== TO-DO LIST ==========")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Completed")
    print("4. Delete Task")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        task_name = input("Enter task name: ")
        task = {"task": task_name, "status": "Pending"}
        tasks.append(task)
        print("Task added successfully!")

    elif choice == "2":
        if not tasks:
            print("No tasks available.")
        else:
            print("\nYour Tasks:")
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task['task']} - {task['status']}")

    elif choice == "3":
        if not tasks:
            print("No tasks available to mark complete.")
        else:
            print("\nTasks:")
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task['task']} - {task['status']}")
            try:
                task_num = int(input("Enter task number to mark as completed: "))
                if 1 <= task_num <= len(tasks):
                    tasks[task_num - 1]["status"] = "Completed"
                    print("Task marked as completed!")
                else:
                    print("Invalid task number.")
            except ValueError:
                print("Please enter a valid number.")

    elif choice == "4":
        if not tasks:
            print("No tasks available to delete.")
        else:
            print("\nTasks:")
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task['task']} - {task['status']}")
            try:
                task_num = int(input("Enter task number to delete: "))
                if 1 <= task_num <= len(tasks):
                    removed_task = tasks.pop(task_num - 1)
                    print(f"Task '{removed_task['task']}' deleted successfully!")
                else:
                    print("Invalid task number.")
            except ValueError:
                print("Please enter a valid number.")

    elif choice == "5":
        print("Thank you for using To-Do List!")
        break

    else:
        print("Invalid choice! Please try again.")
````
